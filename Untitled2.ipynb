```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

distance = ctrl.Antecedent(np.arange(0, 51, 1), 'distance')
traffic = ctrl.Antecedent(np.arange(0, 101, 1), 'traffic')
demand = ctrl.Antecedent(np.arange(0, 101, 1), 'demand')
price = ctrl.Consequent(np.arange(0, 101, 1), 'price')

distance['short'] = fuzz.trapmf(distance.universe, [0, 0, 2, 3])
distance['medium'] = fuzz.trimf(distance.universe, [2, 5, 8])
distance['long'] = fuzz.trimf(distance.universe, [6, 13, 20])
distance['very_long'] = fuzz.trapmf(distance.universe, [15, 30, 50, 50])

traffic['low'] = fuzz.trimf(traffic.universe, [0, 0, 30])
traffic['medium'] = fuzz.trimf(traffic.universe, [20, 50, 70])
traffic['high'] = fuzz.trimf(traffic.universe, [60, 100, 100])

demand['low'] = fuzz.trimf(demand.universe, [0, 0, 30])
demand['medium'] = fuzz.trimf(demand.universe, [20, 50, 70])
demand['high'] = fuzz.trimf(demand.universe, [60, 100, 100])

price['low'] = fuzz.trimf(price.universe, [0, 0, 30])
price['medium'] = fuzz.trimf(price.universe, [30, 50, 70])
price['high'] = fuzz.trimf(price.universe, [60, 80, 90])
price['very_high'] = fuzz.trimf(price.universe, [80, 100, 100])

rule1 = ctrl.Rule(distance['short'] & traffic['low'] & demand['low'], price['low'])
rule2 = ctrl.Rule(distance['short'] & traffic['medium'] & demand['high'], price['medium'])
rule3 = ctrl.Rule(distance['medium'] & traffic['medium'] & demand['high'], price['high'])
rule6 = ctrl.Rule(distance['very_long'] & traffic['high'] & demand['high'], price['very_high'])

pricing_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule6])
pricing_sim = ctrl.ControlSystemSimulation(pricing_ctrl)

pricing_sim.input['distance'] = 18
pricing_sim.input['traffic'] = 85
pricing_sim.input['demand'] = 90

pricing_sim.compute()

print(f"Mức giá đề xuất (tỷ lệ): {pricing_sim.output['price']:.2f}%")
price.view(sim=pricing_sim)
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

rating = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'rating')
sales_volume = ctrl.Antecedent(np.arange(0, 1001, 1), 'sales_volume')
profit_margin = ctrl.Antecedent(np.arange(0, 101, 1), 'profit_margin')
seasonal_event = ctrl.Antecedent(np.arange(0, 101, 1), 'seasonal_event')
competitor_discounts = ctrl.Antecedent(np.arange(0, 101, 1), 'competitor_discounts')

discount = ctrl.Consequent(np.arange(0, 71, 1), 'discount')

rating['low'] = fuzz.trapmf(rating.universe, [0, 0, 3.8, 4.2])
rating['medium'] = fuzz.trimf(rating.universe, [4.0, 4.4, 4.7])
rating['high'] = fuzz.trapmf(rating.universe, [4.5, 4.8, 5, 5])

sales_volume['low'] = fuzz.trapmf(sales_volume.universe, [0, 0, 300, 500])
sales_volume['medium'] = fuzz.trimf(sales_volume.universe, [400, 550, 700])
sales_volume['high'] = fuzz.trapmf(sales_volume.universe, [600, 850, 1000, 1000])

profit_margin['low'] = fuzz.trapmf(profit_margin.universe, [0, 0, 35, 45])
profit_margin['medium'] = fuzz.trimf(profit_margin.universe, [40, 50, 60])
profit_margin['high'] = fuzz.trapmf(profit_margin.universe, [55, 75, 100, 100])

seasonal_event['none'] = fuzz.trapmf(seasonal_event.universe, [0, 0, 25, 45])
seasonal_event['medium'] = fuzz.trimf(seasonal_event.universe, [35, 55, 75])
seasonal_event['high'] = fuzz.trapmf(seasonal_event.universe, [65, 85, 100, 100])

competitor_discounts['low'] = fuzz.trapmf(competitor_discounts.universe, [0, 0, 30, 50])
competitor_discounts['medium'] = fuzz.trimf(competitor_discounts.universe, [40, 60, 80])
competitor_discounts['high'] = fuzz.trapmf(competitor_discounts.universe, [70, 90, 100, 100])

discount['very_low'] = fuzz.trimf(discount.universe, [0, 2, 5])
discount['low'] = fuzz.trimf(discount.universe, [5, 12, 20])
discount['medium'] = fuzz.trimf(discount.universe, [15, 25, 35])
discount['high'] = fuzz.trimf(discount.universe, [30, 45, 60])
discount['very_high'] = fuzz.trimf(discount.universe, [50, 60, 70])

rule1 = ctrl.Rule(rating['high'] & profit_margin['high'], discount['very_low'])
rule2 = ctrl.Rule(rating['low'] & profit_margin['high'], discount['medium'])
rule3 = ctrl.Rule(seasonal_event['high'] & competitor_discounts['high'], discount['very_high'])
rule4 = ctrl.Rule(rating['medium'] & sales_volume['medium'] & profit_margin['medium'], discount['medium'])
rule5 = ctrl.Rule(profit_margin['low'] & seasonal_event['high'], discount['very_high'])
rule6 = ctrl.Rule(competitor_discounts['low'] & profit_margin['high'], discount['very_low'])
rule7 = ctrl.Rule(rating['low'] & sales_volume['low'], discount['high'])
rule8 = ctrl.Rule(seasonal_event['medium'] | competitor_discounts['medium'], discount['low'])
rule9 = ctrl.Rule(profit_margin['low'] & rating['high'], discount['high'])
rule10 = ctrl.Rule(sales_volume['high'] & profit_margin['medium'], discount['medium'])

discount_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10])
simulation = ctrl.ControlSystemSimulation(discount_ctrl)

simulation.input['rating'] = 4.8
simulation.input['sales_volume'] = 700
simulation.input['profit_margin'] = 40
simulation.input['seasonal_event'] = 10
simulation.input['competitor_discounts'] = 80

simulation.compute()
print(f"Chiết khấu: {simulation.output['discount']:.2f}%")
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

income = ctrl.Antecedent(np.arange(0, 101, 1), 'income')
credit_score = ctrl.Antecedent(np.arange(0, 1001, 1), 'credit_score')
debt_ratio = ctrl.Antecedent(np.arange(0, 101, 1), 'debt_ratio')

risk = ctrl.Consequent(np.arange(0, 101, 1), 'risk')

income['low'] = fuzz.trapmf(income.universe, [0, 0, 15, 30])
income['medium'] = fuzz.trimf(income.universe, [20, 50, 80])
income['high'] = fuzz.trapmf(income.universe, [70, 85, 100, 100])

credit_score['bad'] = fuzz.trapmf(credit_score.universe, [0, 0, 300, 500])
credit_score['average'] = fuzz.trimf(credit_score.universe, [400, 600, 800])
credit_score['good'] = fuzz.trapmf(credit_score.universe, [750, 900, 1000, 1000])

debt_ratio['low'] = fuzz.trapmf(debt_ratio.universe, [0, 0, 20, 40])
debt_ratio['medium'] = fuzz.trimf(debt_ratio.universe, [30, 50, 70])
debt_ratio['high'] = fuzz.trapmf(debt_ratio.universe, [60, 80, 100, 100])

risk['low'] = fuzz.trimf(risk.universe, [0, 20, 40])
risk['medium'] = fuzz.trimf(risk.universe, [35, 50, 70])
risk['high'] = fuzz.trapmf(risk.universe, [60, 80, 100, 100])

rule1 = ctrl.Rule(credit_score['bad'] | debt_ratio['high'], risk['high'])
rule2 = ctrl.Rule(income['high'] & credit_score['good'], risk['low'])
rule3 = ctrl.Rule(income['medium'] & debt_ratio['medium'], risk['medium'])
rule4 = ctrl.Rule(credit_score['average'] | income['low'], risk['medium'])
rule5 = ctrl.Rule(debt_ratio['low'] & credit_score['good'], risk['low'])


risk_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
simulation = ctrl.ControlSystemSimulation(risk_ctrl)

simulation.input['income'] = 45
simulation.input['credit_score'] = 650
simulation.input['debt_ratio'] = 30

simulation.compute()

print(f"Mức độ rủi ro tính toán: {simulation.output['risk']:.2f}%")
risk.view(sim=simulation)
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

input_a = ctrl.Antecedent(np.arange(0, 101, 1), 'input_a')
input_b = ctrl.Antecedent(np.arange(0, 101, 1), 'input_b')
input_c = ctrl.Antecedent(np.arange(0, 101, 1), 'input_c')

result = ctrl.Consequent(np.arange(0, 101, 1), 'result')

input_a['low'] = fuzz.trapmf(input_a.universe, [0, 0, 30, 45])
input_a['medium'] = fuzz.trimf(input_a.universe, [35, 50, 65])
input_a['high'] = fuzz.trapmf(input_a.universe, [55, 75, 100, 100])

input_b['low'] = fuzz.trapmf(input_b.universe, [0, 0, 25, 50])
input_b['medium'] = fuzz.trimf(input_b.universe, [40, 60, 80])
input_b['high'] = fuzz.trapmf(input_b.universe, [70, 85, 100, 100])

input_c['low'] = fuzz.trapmf(input_c.universe, [0, 0, 35, 55])
input_c['medium'] = fuzz.trimf(input_c.universe, [45, 65, 85])
input_c['high'] = fuzz.trapmf(input_c.universe, [75, 90, 100, 100])

result['low'] = fuzz.trimf(result.universe, [0, 25, 50])
result['medium'] = fuzz.trimf(result.universe, [35, 50, 75])
result['high'] = fuzz.trapmf(result.universe, [60, 80, 100, 100])

rule1 = ctrl.Rule(input_a['high'] | input_b['high'], result['high'])
rule2 = ctrl.Rule(input_a['low'] & input_b['low'], result['low'])
rule3 = ctrl.Rule(input_c['high'], result['high'])
rule4 = ctrl.Rule(input_a['medium'] | input_b['medium'] | input_c['medium'], result['medium'])
rule5 = ctrl.Rule(input_c['low'] & input_a['low'], result['low'])

system_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
simulation = ctrl.ControlSystemSimulation(system_ctrl)

simulation.input['input_a'] = 50
simulation.input['input_b'] = 60
simulation.input['input_c'] = 45

simulation.compute()

print(simulation.output['result'])
result.view(sim=simulation)
```
